```python
def milne(f, a, b, h):
    n = math.trunc((b - a) / h)
    if n % 2 != 0: n += 1
    x = a
    sum = 0
    for i in range (1, n + 1):
        x += h
        sum += (h/90)*(7*f(x-h)+32*f(x-h+h/4)+12*f(x-h/2)+32*f(x-h/4)+7*f(x))
    return sum
```

```python
def milneErr(f, a, b, p, eps, h):
    cnt = 0
    Ih = milne(f, a, b, h)
    while True:
        h /= 2
        cnt += 1
        Ih2 = milne(f, a, b, h)
        if abs(Ih2 - Ih) / (2 ** p - 1) < eps:
            break
        Ih = Ih2
    return Ih2, h, cnt

p = 4
k = 9
eps = p * (10 ** (-k))
h = b - a
Ik = milneErr(f, a, b, p, eps, h)
#print(eps, J, Ik, abs(J - Ik[0]))
```
